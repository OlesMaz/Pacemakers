```python
import pandas as pd
data = pd.read_csv('Euro_2012_stats_TEAM.csv')
data.columns
```

### Question 1

#### How many(%) fouls of each country led to yellow card?

```python
pd.concat([data["Team"], data["Yellow Cards"] / data["Fouls Conceded"] * 100], axis=1).rename(columns={0:"Yellow card(%)"}).sort_values(by="Yellow card(%)",ascending=False)

```

### Question 2

#### Shots of which country were blocked the most?

```python
data["Blocked_shots"] = data["Total shots (inc. Blocked)"] - data["Shots off target"] - data["Shots on target"] - data["Hit Woodwork"]
data.loc[:, ["Team", "Blocked_shots"]].sort_values(by="Blocked_shots", ascending=False)
```

### Question 3

#### Which team used the most players?

```python
players_used = data[['Team', 'Players Used']].sort_values(by = "Players Used", ascending = False).reset_index().head(10)
players_used
```

### Question 4

#### How many goals were scored and which team scored the most?

```python
total_goals = data['Goals'].sum()
print(f"{total_goals} goals were scored in total")
```

```python
max_goals = data["Goals"].idxmax()
print(f'{data.loc[max_goals, "Team"]} Scored the most goals')
```

### Question 5

#### Which teams have best goalkeepers?

```python
best_goalkeepers = data[['Team', 'Saves-to-shots ratio']].loc[data["Clean Sheets"] > 1]
best_goalkeepers.sort_values(['Saves-to-shots ratio'], ascending = False).head(10)
```

### Question 6

#### What is % of dribbling out of all touches in each team?

```python
dribbling = data
dribbling.groupby("Team").apply(lambda data: round((data["Dribbles"] / (data["Touches"]) * 100).mean(), 2)).reset_index().rename(columns={0:"Dribbling %"}).sort_values(by="Dribbling %", ascending=False)
```

### Question 7

#### Teams with best Goal Difference?

```python
data["Difference"] = data["Goals"] - data["Goals conceded"]
gd = data[["Team","Difference"]].sort_values(by = "Difference", ascending=False).head(10)
gd
```

### Question 8

#### What type of goal each team scored more: Penalty or Headed?

```python
data["Pen or Head"] = data.apply(lambda row: "Pen" if row["Penalty goals"] > row["Headed goals"] else "Head", axis="columns")
data[["Team", "Pen or Head"]]
```

### Question 9

#### How many passes(%) led to offside?

```python
data["Offside passes %"] = data["Offsides"] / data["Passes"] * 100
data[["Team", "Offside passes %"]].sort_values(by="Offside passes %", ascending=False)
```

### Question 10

#### How much is the number of corners taken related to goals?

```python
data["Corners_levels"] = data.apply(lambda row: "Low" if row["Corners Taken"] <= data["Corners Taken"].describe().loc["25%"] else "Med-Low" if row["Corners Taken"] <= data["Corners Taken"].describe().loc["50%"] else "Med-High" if row["Corners Taken"] <= data["Corners Taken"].describe().loc["75%"] else "High", axis='columns')
data["Goals_levels"] = data.apply(lambda row: "Low" if row["Goals"] <= data["Goals"].describe().loc["25%"] else "Med-Low" if row["Goals"] <= data["Goals"].describe().loc["50%"] else "Med-High" if row["Goals"] <= data["Goals"].describe().loc["75%"] else "High", axis='columns')
relation_lvl = data.apply(lambda row: row["Corners_levels"] == row["Goals_levels"], axis="columns").sum() / data.Team.count() * 100
print(f"They are related in {round(relation_lvl,2)}% of teams")
```
